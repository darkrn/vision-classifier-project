```python
import torch
import torchvision
import torchvision.transforms as transforms

# 1. Define baseline transformations (Convert to tensor and scale/normalize)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# 2. Load the local data
train_set = torchvision.datasets.CIFAR10(root='../data', train=True, download=True, transform=transform)
test_set = torchvision.datasets.CIFAR10(root='../data', train=False, download=True, transform=transform)

# 3. Create a 10% validation split for tracking performance
train_size = int(0.9 * len(train_set))
val_size = len(train_set) - train_size
train_subset, val_subset = torch.utils.data.random_split(train_set, [train_size, val_size])

print(f"Dataset splits created successfully!")
print(f"Training samples: {len(train_subset)}")
print(f"Validation samples: {len(val_subset)}")
print(f"Testing samples: {len(test_set)}")

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def imshow(img):
    img = img / 2 + 0.5  # unnormalize back to [0, 1]
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Create a small loader to grab a few sample images
preview_loader = torch.utils.data.DataLoader(train_subset, batch_size=4, shuffle=True)
dataiter = iter(preview_loader)
images, labels = next(dataiter)

# Show images and print their target labels
imshow(torchvision.utils.make_grid(images))
print('Labels: ', ' '.join(f'{classes[labels[j]]:5s}' for j in range(4)))
```

```python
import torch.nn as nn
import torch.nn.functional as F

class BaselineCNN(nn.Module):
    def __init__(self):
        super().__init__()
        # Conv layer 1: 3 color channels input -> 6 feature maps output. 5x5 kernel.
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        # Conv layer 2: 6 feature maps input -> 16 feature maps output. 5x5 kernel.
        self.conv2 = nn.Conv2d(6, 16, 5)
        
        # Fully connected layers (16 feature maps * 5x5 image resolution remaining)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10) # 10 outputs matching CIFAR-10 classes

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1) # Flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Automatically leverage a GPU (CUDA) if your system has one available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = BaselineCNN().to(device)
print(f"Model initialized and loaded onto device: {device}")
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# Build official batch loaders
train_loader = torch.utils.data.DataLoader(train_subset, batch_size=32, shuffle=True)

print("Starting Baseline Training Loop...")
for epoch in range(2):
    running_loss = 0.0
    model.train()
    
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)

        # Zero out parameter gradients from the previous step
        optimizer.zero_grad()

        # Forward pass + Backward Pass + Weight Update
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        if i % 200 == 199:  # Print updates every 200 batches
            print(f'Epoch [{epoch + 1}/2], Batch [{i + 1}/{len(train_loader)}], Loss: {running_loss / 200:.3f}')
            running_loss = 0.0

print('Finished Baseline Training!')

# Save model weights to your models folder
torch.save(model.state_dict(), '../models/baseline_cnn.pt')
print("Saved baseline checkpoint to: ../models/baseline_cnn.pt")
```

```python
correct = 0
total = 0
val_loader = torch.utils.data.DataLoader(val_subset, batch_size=32, shuffle=False)

# Switch model to evaluation mode
model.eval()

with torch.no_grad(): # No need to calculate gradients during evaluation
    for data in val_loader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model(images)
        
        # The class with the highest energy is what we choose as prediction
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

baseline_accuracy = 100 * correct / total
print(f'Accuracy of the baseline network on the 5000 validation images: {baseline_accuracy:.2f}%')
```
